```python
from sklearn.datasets import make_classification, make_regression
from sklearn.model_selection import train_test_split

# Create classification dataset (500 rows)
X_class, y_class = make_classification(
    n_samples=500,
    n_features=10,
    n_informative=8,
    n_redundant=2,
    n_classes=2,
    random_state=42,
    flip_y=0.1
)

# Create regression dataset (500 rows)
X_reg, y_reg = make_regression(
    n_samples=500,
    n_features=10,
    n_informative=8,
    random_state=42,
    noise=10
)

# Save datasets
import pandas as pd
pd.DataFrame(X_class, columns=[f'feature_{i}' for i in range(10)]).assign(target=y_class).to_csv('classification_data.csv', index=False)
pd.DataFrame(X_reg, columns=[f'feature_{i}' for i in range(10)]).assign(target=y_reg).to_csv('regression_data.csv', index=False)

print("Datasets created successfully!")
print(f"Classification dataset shape: {X_class.shape}, Target classes: {np.unique(y_class)}")
print(f"Regression dataset shape: {X_reg.shape}, Target range: [{y_reg.min():.2f}, {y_reg.max():.2f}]")
```
